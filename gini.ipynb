```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
econ=pd.read_csv('Global Economy Indicators.csv')#,sep='|',encoding='cp1252'
col2=[]
# cleaning the names of the columns
for c in econ.columns:
    col2.append(c.strip())
econ.columns=col2
# cleaning the countries
for i, c in econ['Country'].items():
    econ.iloc[i,1]=c.strip()
```

```python
#data on freedom index, GNI per capita, GINI index, etc
gini=pd.read_csv('freedom.csv')
gini
```

```python
#geographic data
geo=pd.read_csv('dimgeo.csv')
geo
```
