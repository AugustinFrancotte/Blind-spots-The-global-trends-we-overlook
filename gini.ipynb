```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
econ=pd.read_csv('datasets\Global Economy Indicators.csv')#,sep='|',encoding='cp1252'
col2=[]
# cleaning the names of the columns
for c in econ.columns:
    col2.append(c.strip())
econ.columns=col2
# cleaning the countries
for i, c in econ['Country'].items():
    econ.iloc[i,1]=c.strip()
```

```python
#data on freedom index, GNI per capita, GINI index, etc
gini=pd.read_csv('datasets\\freedom.csv')
gini
```

```python
#geographic data
geo=pd.read_csv('datasets\dimgeo.csv')
geo
```

```python
gini2=pd.merge(gini,geo,on='Country',how='inner')
```

```python
gini2
```

```python
pays=['China','India','United States', 'France', 'Switzerland','Germany','Algeria','Morocco','Taiwan','Japan','Sweden','Denmark',
      'Malaysia','South Korea','Poland','Russia','Brazil','Argentina','Columbia','Chile','South Africa','Italy','Spain','Nigeria',
     'Cambodia','Thailand','Indonesia','Iran','Canada','Mexico','Singapour','Hong Kong','Taiwan','Egypt','Liberia','Angola',
     'Slovenia','Albania','Greece','Austria']
```

```python
plt.figure(figsize=(15, 12))
sns.regplot(x="Freedom Idx/Gini Idx", y="GniPerCapitaUSD", data=gini2)
sns.scatterplot(data=gini2, x='Freedom Idx/Gini Idx', y='GniPerCapitaUSD',
                hue='Continental region', style='Continental region',s=100)

#plt.legend()
plt.ylim(-2000,122000)

for i, row in gini2.iterrows():
    if row['Country'] in pays:  # Annotation uniquement si le label n'est pas vide
        plt.text(row['Freedom Idx/Gini Idx'] + 0.01, row['GniPerCapitaUSD'], row['Country'], fontsize=12, color='black')
plt.savefig('graphs\\Gini-freedom-gnipercapita.png',bbox_inches='tight')
#but : ajouter le nom de quelques pays, et la rÃ©gion / sub rÃ©gion en couleurs
```
