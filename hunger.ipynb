```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
econ2=pd.read_csv('datasets\Global Economy Indicators.csv')#,sep='|',encoding='cp1252'
col22=[]

# cleaning the names of the columns
for c in econ2.columns:
    col22.append(c.strip())
econ2.columns=col22
# cleaning the countries
for i, c in econ2['Country'].items():
    econ2.iloc[i,1]=c.strip()
```

```python
hun=pd.read_csv('datasets\prevalence-of-undernourishment-in-developing-countries-since-1970.csv')
```

```python
col=list(hun.columns)
col[-1]='Undernourished people in developping countries(%)'
hun.columns=col
```

```python
sns.lineplot(data=hun,x='Year',y='Undernourished people in developping countries(%)')
plt.title('Evolution of undernourishment in developping countries')
plt.savefig('graphs\\undernourishment.png',bbox_inches='tight')
```
