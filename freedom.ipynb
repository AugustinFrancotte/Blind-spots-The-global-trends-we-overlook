```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import country_converter as coco
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

cc=coco.CountryConverter()
```

```python
econ=pd.read_csv('datasets\Global Economy Indicators.csv')#,sep='|',encoding='cp1252'
col2=[]
# cleaning the names of the columns
for c in econ.columns:
    col2.append(c.strip())
econ.columns=col2
# cleaning the countries
for i, c in econ['Country'].items():
    econ.iloc[i,1]=c.strip()
```

```python
free4=pd.read_excel('datasets\\2024_indexofeconomicfreedom_data.xlsx')#,sep='|',encoding='cp1252'
free4
```

```python
# Extract most recent gini index from wikipedia
gin=pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_income_inequality')
# formal changes in the dataframe
gin[0].columns=['Country','UN Region','WB income group','GINI_WB','y1','GINI_UNU','y2','GINI_OECD','y3','rr']
gin4=gin[0].loc[:,['Country','WB income group','GINI_WB','GINI_UNU']]
gin4=gin4.drop(195) # ! rendre plus robuste
gin4['CountryID']=cc.pandas_convert(series=gin4['Country'],to='ISO3')
```

```python
gin4
```

```python
# extract most recent GNI per capita
percap=pd.read_csv('datasets\\gni-per-capita-by-country-2024.csv')
percap['CountryID']=cc.pandas_convert(series=percap['country'],to='ISO3')
```

```python
percap
```

```python
# extract most reccent incarceration rate by country 
prison=pd.read_csv("datasets\incarceration-rates-by-country-2025.csv")
coll=list(prison.columns)
coll[1]='Country'
prison.columns=coll
prison['CountryID']=cc.pandas_convert(series=prison['Country'],to='ISO3')
prison
```

```python
# extract most recent freedom index per country from wikipedia
scrap2=pd.read_html("http://en.wikipedia.org/wiki/Index_of_Economic_Freedom")
f2021=scrap2[0].loc[:,['Country','Score']]
f2021.columns=['Country','Freedom Index']
f2021['CountryID']=cc.pandas_convert(series=f2021['Country'],to='ISO3')
```

```python
# merging the multiple data into one table containing : 
merged=pd.merge(percap,prison, how='outer',on=['CountryID'])
merged=pd.merge(merged,gin4, how='outer',on=['CountryID'])
merged=pd.merge(merged,f2021, how='outer',on=['CountryID'])
```

```python
# new column : Freedom index / GINI index
# This column measure how much a country is free (Freedom index) and how cooperative the country is (GINI index)
merged['Freedom Idx/Gini Idx']=merged['Freedom Index']/merged['GINI_UNU']
```

```python
# making sure every float column is actually float
merged['Freedom Index']=merged['Freedom Index'].astype('float')
merged['GINI_WB']=merged['GINI_WB'].astype('float')
merged['GNIPerCapita']=merged['GNIPerCapita'].astype('float')
merged['GNIPerCapitaUSD']=merged['GNIPerCapitaUSD'].astype('float')
#merged['IncarcerationRates_IncarcerationRatePer100k_rate_MultiYear']=merged[''].str.replace(',', '').astype('float')
#merged['Imprisonment']=merged['Imprisonment'].str.replace(',', '').astype('float')
merged['Freedom Idx/Gini Idx']=merged['Freedom Idx/Gini Idx'].astype('float')
merged['GINI Index(UNU)']=merged['GINI Index(UNU)'].astype('float')
```

```python
# dropping doubles
merged=merged.drop(columns=['country','Country_x','Country_y','GniPerCapita_RecentYear2022DY'])
```

```python
merged.columns=['GNIPerCapita','GNIPerCapitaUSD','CountryID','Imprisonment Rate (per 100K people)','Imprisonment','WB Income Group','GINI Index(WB)','GINI Index(UNU)','Country','Freedom Index','Freedom Idx/Gini Idx' ]
```

```python
merged.to_csv('freedom.csv')
```
