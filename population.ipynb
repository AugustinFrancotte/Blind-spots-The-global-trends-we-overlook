```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop=pd.read_csv('datasets\population-regions-with-projections.csv')  
```

```python
continents = ["Africa", "Asia", "Europe", "Oceania", "North America", "South America"]
sns.set_theme()
# Filter data between 1950 and 2025
df_filtered = pop[(pop["Entity"].isin(continents)) & (pop["Year"] >= 1950) & (pop["Year"] <= 2025)]

# organize data on the graph
df_pivot = df_filtered.pivot(index="Year", columns="Entity", values="Population")
df_pivot_billion = df_pivot / 1e9  # Conversion in billions

# Cumulative plot
plt.figure(figsize=(12, 6))
plt.stackplot(df_pivot_billion.index, 
              [df_pivot_billion[c] for c in continents], 
              labels=continents)

plt.title("World Population by Continent (1950–2025)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Population (billions)")
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig('graphs\\population.png',bbox_inches='tight')
plt.show()
```
