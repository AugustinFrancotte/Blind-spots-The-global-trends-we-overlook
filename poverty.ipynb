```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
econ2=pd.read_csv('datasets\Global Economy Indicators.csv')#,sep='|',encoding='cp1252'
col22=[]

# cleaning the names of the columns
for c in econ2.columns:
    col22.append(c.strip())
econ2.columns=col22
# cleaning the countries
for i, c in econ2['Country'].items():
    econ2.iloc[i,1]=c.strip()
```

```python
pov=pd.read_csv('datasets\poverty.csv')
pov['Indicator Name'].unique()
```

```python
pov=pov.rename(columns={'Country Code':'CountryID'})
```

```python
pov685=pov[pov['Indicator Name']=='Poverty headcount ratio at $6.85 a day (2017 PPP) (% of population)']
```

```python

# Selection of the thresholds of poverty
poverty_indicators = {
    "2.15$ per day": "SI.POV.DDAY",
    "3.65$ per day": "SI.POV.LMIC",
    "6.85$ per day": "SI.POV.UMIC",
}

# Filter world data
global_poverty = pov[pov["Country Name"] == "World"]

# Transform the dataset by year
years = [col for col in pov.columns if col.isdigit()]
global_poverty_data = global_poverty.set_index("Indicator Code")[years].T

# Convert the index and years in integers
global_poverty_data.index.name = "Year"  # Renommer pour plus de clarté
global_poverty_data.index = global_poverty_data.index.astype(int)


# Draw the Evolution of % for each threshold
plt.figure(figsize=(10, 5))
for label, code in poverty_indicators.items():
    plt.plot(global_poverty_data.index, global_poverty_data[code], marker="o", label=label)

# Graph features
plt.title("Evolution of the percentage of population living in poverty")
plt.xlabel("Year")
plt.ylabel("Percentage of global population")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.savefig('graphs\\poverty.png',bbox_inches='tight')
plt.show()
```
