```python

# Sélection des seuils de pauvreté pertinents
poverty_indicators = {
    "2.15$ par jour": "SI.POV.DDAY",
    "3.65$ par jour": "SI.POV.LMIC",
    "6.85$ par jour": "SI.POV.UMIC",
}

# Filtrer les données mondiales
global_poverty = pov[pov["Country Name"] == "World"]

# Transformer le dataset : mise en forme des années
years = [col for col in pov.columns if col.isdigit()]
global_poverty_data = global_poverty.set_index("Indicator Code")[years].T

# Convertir l'index (les années) en entiers
global_poverty_data.index.name = "Year"  # Renommer pour plus de clarté
global_poverty_data.index = global_poverty_data.index.astype(int)


# Tracer l'évolution du % sous chaque seuil
plt.figure(figsize=(10, 5))
for label, code in poverty_indicators.items():
    plt.plot(global_poverty_data.index, global_poverty_data[code], marker="o", label=label)

# Mise en forme du graphique
plt.title("Évolution du % d'humains vivant sous les seuils de pauvreté")
plt.xlabel("Année")
plt.ylabel("Pourcentage de la population mondiale")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)


plt.show()
```
